```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
miles_df = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv',parse_dates = [0])
```

```python
miles_df.head()
```

```python
miles_df.info()
```

#### Downsampling - Reduce the number of Samples reducing the frequency of data. 
Ex: MOnthly to Quartlery, Quarterly Data into Yearly Data.

```python
quarterly_miles_df = miles_df.resample('Q', on='Month').mean()
quarterly_miles_df.head()
```

```python
yearly_miles_df = miles_df.resample('Y', on ='Month').sum()
yearly_miles_df.head()
```

#### Upsampling - Increasing the Frequency of your data.
Adding Samples to data to increase the frequency is called Upsampling.
Ex: Yearly Data into Quarterly, Quarterly into Monthly.

```python
daily_miles_df = miles_df.resample('D',on = 'Month').mean()
daily_miles_df.head(35)
```

```python
interpolated_daily_miles_df = daily_miles_df.interpolate(method = 'linear')
interpolated_daily_miles_df.head(35)
```

```python
interpolated_daily_miles_df.plot()
plt.show()
```

```python
miles_df['MilesMM'].plot()
```

```python
interpolated_daily_miles_df1 = daily_miles_df.interpolate(method = 'spline', order = 2)
interpolated_daily_miles_df1.head(35)
```

```python
figure, axa=plt.subplots(1,2,figsize=(16,4))
interpolated_daily_miles_df1.plot(ax = axa[0] )
interpolated_daily_miles_df.plot(color='r', ax = axa[1])
axa[0].set_title('Linear Intepolated values having Sharper Edges')
axa[1].set_title('Polynomial (2) Intepolated values having Curved Edges')
plt.show()
```
